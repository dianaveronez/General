```python
# Setup plot
fig, ax = plt.subplots(figsize=(12, 7), subplot_kw={'projection': ccrs.PlateCarree()})
cbar_ax = fig.add_axes([0.92, 0.25, 0.02, 0.5])  # [left, bottom, width, height]

# Pre-load one file to get extent and color scale
ds0 = xr.open_dataset(files[0])
lon = ds0.longitude
lat = ds0.latitude
vmin = 0
vmax = max(
    float(xr.open_dataset(f)['APCP_surface'].max().values)
    for f in files
)

# Initial plot
mesh = ax.pcolormesh(lon, lat, ds0['APCP_surface'].squeeze(), cmap='gist_ncar_r', vmin=vmin, vmax=vmax, transform=ccrs.PlateCarree())
cbar = plt.colorbar(mesh, cax=cbar_ax, label='Precipitation (mm)', orientation='vertical')

def update(i):
    ds = xr.open_dataset(files[i])
    data = ds['APCP_surface'].squeeze()

    ax.clear()
    ax.set_title(f"{str(ds.time.values)}", fontsize=14, fontweight='bold')
    ax.set_extent([lon.min(), lon.max(), lat.min(), lat.max()])
    ax.add_feature(cfeature.COASTLINE)
    ax.add_feature(cfeature.BORDERS, linewidth=0.5)
    ax.add_feature(cfeature.STATES, linewidth=0.3)

    mesh = ax.pcolormesh(lon, lat, data, cmap='gist_ncar_r', vmin=vmin, vmax=vmax, transform=ccrs.PlateCarree())
    return [mesh]

ani = animation.FuncAnimation(fig, update, frames=len(files), blit=False)

# Save as GIF or MP4
ani.save(r"C:\Users\dveronez\George Mason University - O365 Production\FloodHazardsResearchLab-GRP - Documents\04_Projects\2024_Project_NOAA_EESLR_2\5_GIS_Datasets\GMU_new_data\Precipitation\2012\aorc_rainfallD.gif", writer='pillow', fps=5, dpi=200)
```
